# 4. Analysis of Data

Conditional statements, looping, groupby, using NumPy, dictionary or lists and defining a custom function to create reuseable code.  

```python
#Reusable code, looping, conditional statements
import regions


df = pd.read_csv(r'../data/Property Dataset Final.csv')
d = df.to_dict('list')
d['Province'] = []

for pc in d['PropertyCounty']:
    try:
        is_match = [c for c in regions.County if c.value[0] in pc]
        if is_match:
            province_name = is_match[0].value[1].value
            d['Province'].append(province_name)
        else:
            d['Province'].append('Invalid Data')
    except Exception as exc:
        d['Province'].append('Invalid Data')

new_df = pd.DataFrame.from_dict(d)
new_df.head(100)  
```

```python
#Dropping the columns with data that I will not use
df3 = new_df.drop(['PropertyID','PropertyStreetCode','PropertyCurrentStatus','PropertyType','PropertyOwningBody','PropertySubType'], axis = 1)
df3.rename(columns ={'PropertyLocalAuthorityArea': 'LocalAuthority', 'PropertyAreaOffice':'HousingManager'}, inplace=True)
df3.tail()
```

```python
#filling in blank data in rows
df3.isna().sum()
```

```python
#filling in blank data in rows
df3['PropertyTown'].fillna('To be updated', inplace = True)
df3['PropertyCounty'].fillna('To be updated', inplace = True)
df3['LocalAuthority'].fillna('To be updated', inplace = True)
df3['PropertyStyle'].fillna('To be updated', inplace = True)
df3.isna().sum()
```

```python
df3.info()
#checking how many columns are str
```

```python
#cleaning out white spaces from str data
df3['PropertyTown'] = df3['PropertyTown'].str.strip()
df3['PropertyCounty'] = df3['PropertyCounty'].str.strip()
df3['HousingManager'] = df3['HousingManager'].str.strip()
df3['LocalAuthority'] = df3['LocalAuthority'].str.strip()
df3['PropertyStyle'] = df3['PropertyStyle'].str.strip()
df3['Province'] = df3['Province'].str.strip()
```

```python
#removing data from before 1998 and after end of 2022
df3 = df3[(df3['YearOfConstruction'] >= 1998)& (df3['YearOfConstruction'] <= 2022)]
df3.shape
```

```python
#showing construction of delivered homes from 1998 - 2022
x = df3.hist(column=('YearOfConstruction'), bins=25, grid=False, figsize=(12,8), color = 'green')
```

```python
#removing invalid data
df3.drop(df3[df3['HousingManager'] =='Sold/Disposed'].index, inplace = True)
df3.HousingManager.value_counts().sort_index()
```

```python
#removing invalid data
df3.drop(df3[df3['Province'] =='Invalid Data'].index, inplace = True)
df3.Province.value_counts().sort_index()
```

```python
#removing outliers as no volume of this prooerty type delivered during 1998-2022
df3.drop(df3[df3['PropertyStyle'] =='Bedsit'].index, inplace = True)
df3.drop(df3[df3['PropertyStyle'] =='Bungalow'].index, inplace = True)
df3.drop(df3[df3['PropertyStyle'] =='Offices'].index, inplace = True)
df3.drop(df3[df3['PropertyStyle'] =='Guest room'].index, inplace = True)
df3.drop(df3[df3['PropertyStyle'] =='Room in Group Home'].index, inplace = True)
df3.drop(df3[df3['PropertyStyle'] =='To be updated'].index, inplace = True)
df3.Province.value_counts().sort_index()
```

```python
#Which years delivered the highest volume of homes and displaying numbers as whole numbers
pd.options.display.float_format = "{:.0f}".format
h = df3.describe()
h
```
